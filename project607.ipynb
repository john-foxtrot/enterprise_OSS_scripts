```python
import os
import requests
import json
import numpy as numpy
import pandas as pd

import requests
from requests.auth import HTTPBasicAuth
```

```python
username = 'square'
authentication = HTTPBasicAuth(username,'2a9d3467ff3e8e0b5e37b73ab95e2a2eeaa314cd')

data = requests.get('https://api.github.com/users/' + username, auth = authentication)
data = data.json()
#print(data)  #print_my_repository_data
```

```python
print(data['repos_url'])
```

```python
url = data['repos_url']
page_no = 1
repos_data = []

while(True):
    response = requests.get(url,auth = authentication)
    response = response.json()
    repos_data = repos_data + response
    repos_fetched = len(response)
    print("total repos fetched: {}".format(repos_fetched) )
    if (repos_fetched == 30):
        page_no = page_no + 1
        url = data['repos_url'] + '?page=' + str(page_no)
    else:
        break

#print(repos_data[0]) # repository_data_fetching
```

```python
repos_data[0]
```

```python
repos_information = []
for i,repo in enumerate(repos_data):
    dat = []
    dat.append(repo['id'])
    dat.append(repo['name'])
    dat.append(repo['owner']['type'])
    dat.append(repo['description'])
    dat.append(repo['created_at'])
    dat.append(repo['updated_at'])
    dat.append(repo['owner']['login'])
    dat.append(repo['license']['name'] if repo['license']!= None else None)
    dat.append(repo['has_wiki'])
    dat.append(repo['forks_count'])
    dat.append(repo['open_issues_count'])
    dat.append(repo['stargazers_count'])
    dat.append(repo['watchers_count'])
    dat.append(repo['url'])
    dat.append(repo['commits_url'].split("{")[0])
    dat.append(repo['url'] + '/languages')
    repos_information.append(dat)
```

```python
repos_df = pd.DataFrame(repos_information,columns = ['Id','Name','Type','Description', 'Created on', 'Updated on', 
                                                      'Owner', 'License', 'Includes wiki', 'Forks count', 
                                                      'Issues count', 'Stars count', 'Watchers count',
                                                      'Repo URL', 'Commits URL', 'Languages URL'])
repos_df.head(10)
#repos_df
```

```python
for i in range(repos_df.shape[0]):
    response = requests.get(repos_df.loc[i,'Languages URL'],auth = authentication)
    response = response.json()
    print(i , response)
    if response != {}:
        languages = []
        for key, value in response.items():
            languages.append(key)
        languages = ', '.join(languages)
        repos_df.loc[i, 'Languages'] = languages
    else:
        repos_df.loc[i, 'Languages'] = ""
            
```

```python
repos_df.to_csv('repos_info.csv', index = False)
```

```python
response = requests.get(repos_df.loc[0,'Commits URL'], auth = authentication)
response.json()                                    
```

```python
commits_info = []
for i in range(repos_df.shape[0]):
    url = repos_df.loc[i,'Commits URL']
    page_no = 1
    while(True):
        response = requests.get(url, auth = authentication)
        response = response.json()
        print("URL: {}, commits{}".format(url,len(response)))
        for commit in response:
            commit_data = []
            commit_data.append(repos_df.loc[i,'Id'])
            commit_data.append(commit['sha'])
            commit_data.append(commit['commit']['committer']['date'])
            commit_data.append(commit['commit']['committer']['email'])
            commit_data.append(commit['commit']['message'])
            commits_info.append(commit_data)
        
        if(len(response) == 30):
            page_no = page_no + 1
            url = repos_df.loc[i,'Commits URL'] + '?page=' + str(page_no)
        else:
            break
            
```

```python
commits_df = pd.DataFrame(commits_info, columns = ['Repo Id', 'Commit Id', 'Date','committer', 'Message'])
commits_df.head(5)
```

```python
commits_df.to_csv('commits_info.csv', index = False)
```
